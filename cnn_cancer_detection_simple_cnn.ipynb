```python

import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import utils
import os
```

```python
training_labels = utils.load_training_labels(root_dir, num_samples=40000, balanced=True)
X = utils.get_all_training_images(training_labels, root_dir)
y = training_labels['label']
print(X.shape)
print(y.shape)
```

```python
model = keras.models.Sequential([
    keras.layers.Input(shape=(96, 96, 3)),
    keras.layers.Rescaling(1./255),
    keras.layers.Cropping2D(16),
    keras.layers.Conv2D(32, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(64, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Conv2D(128, (3,3), activation='relu'),
    keras.layers.MaxPooling2D((2,2)),
    keras.layers.Flatten(),
    keras.layers.Dense(128, activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.0001), loss='binary_crossentropy', metrics=['BinaryAccuracy'])
model.summary()
```

```python
checkpointer = keras.callbacks.ModelCheckpoint(
    filepath=os.path.join(root_dir, 'model.h5.best_weights.keras'),
    verbose=1,
    save_best_only=True,
    monitor='val_BinaryAccuracy',
    mode='max'
)

history = model.fit(X, y, epochs=100, batch_size=256, validation_split=0.2, callbacks=[checkpointer])
```
